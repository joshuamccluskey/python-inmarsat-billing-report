# FY21 Concanted PI Type Files Collection 1


```python
import os
import glob
import pandas as pd
os.chdir("./FY 21 PI Type Revised")
```

```python
extension = 'xlsx'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
#combine all files in the list
combined_excel = pd.concat([pd.read_excel(f) for f in all_filenames ])
#export to csv
combined_excel.to_excel( "all_data_pi_type collection 1.xlsx", index=False)
```

```python
df = pd.read_excel('all_data_pi_type collection 1.xlsx')
df = df.groupby(['PI','Type']).sum()
df = df.drop(columns=['IDs'])
df['FY 21 Collection 1 Totals'] = df.iloc[:, 0:14].sum(axis=1)
df.to_excel('FY 21 INMARSAT Collection 1 Totals PI Type.xlsx')
df
```
