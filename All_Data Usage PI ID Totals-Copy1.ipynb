```python
import pandas as pd
import re
import os
```

```python
files = [file for file in os.listdir('./FY 21 Data')]
all_months_data = pd.DataFrame()
for file in files:
    df = pd.read_excel('./FY 21 Data/'+file)
    all_months_data = pd.concat([all_months_data, df])
all_months_data.to_excel('all_data_ID_Totals.xlsx', index=False)
all_months_data
```

```python
all_months_data_groups = all_months_data.groupby(['PI','IDs']).sum()
all_months_data_groups.to_csv('all_data_ID_Totals_Grouped.csv')
all_months_data_groups 
```

```python

all_months_data_groups['FY 20 Year End Totals'] = all_months_data_groups["Apr'20 Due"] + all_months_data_groups["Aug'20 Due"] + all_months_data_groups["Dec'19 Due"] + all_months_data_groups["Feb'20 Due"] + all_months_data_groups["Jan'20 Due"] + all_months_data_groups["Jul'20 Due"] + all_months_data_groups["Jun'20 Due"] + all_months_data_groups["Mar'20 Due"] + all_months_data_groups["May'20 Due"] + all_months_data_groups["Nov'19 Due"] + all_months_data_groups["Oct'19 Due"] + all_months_data_groups["Sep'20 Due"]
all_months_data_groups.to_csv('all_data_ID_Totals_Grouped_Totals.csv')
all_months_data_groups
```
