# FY21 Concanted PI Totals Files Collection 2



```python
import os
import glob
import pandas as pd
os.chdir("./FY 21 PI Totals Collection 2")
```

```python
extension = 'xlsx'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
#combine all files in the list
combined_excel = pd.concat([pd.read_excel(f) for f in all_filenames ])
#export to csv
combined_excel.to_excel( "all_data_pi_totals collection 2.xlsx", index=False)
```

```python
df = pd.read_excel( "all_data_pi_totals collection 2.xlsx")
df = df.drop(columns=['IDs'])
df = df.groupby(['PI']).sum()
df['FY 21 Collection 2 Totals'] = df.iloc[:, 0:14].sum(axis=1)
df.to_excel('FY 21 INMARSAT Collection 2 Totals PI Type.xlsx')
df
```
