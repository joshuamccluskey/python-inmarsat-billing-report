```python
import pandas as pd

import re

import os
```

```python
df = pd.read_csv('all_data_updated_clean.clean.csv')

def get_ID(ID):
    return ID.split(' ')[2]

df['IDs'] = df['ID'].apply(lambda x: get_ID(x))

df['Amounts'] = df['Totals'].apply(lambda x: x.split('$')[1])

df['Amounts'] = pd.to_numeric(df['Amounts'])

df.to_csv('all_data_updated_IDs_Amounts.csv', index=False)

df
```
